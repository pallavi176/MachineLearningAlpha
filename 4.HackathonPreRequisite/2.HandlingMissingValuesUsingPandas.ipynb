## 1. Introduction
--------------------

### What does a Null value look like? 
- Shows up as 'NaN'

### Why is it important to address?
- Models can't take NaN as input

### Types of Null values?
- Numerical: Eg - Number of rooms in the house
- Categorical: Eg - Name of housing society

```python
#df = pd.read_csv('/kaggle/input/bengaluru-house-price-data/Bengaluru_House_Data.csv')
df = pd.read_csv('Bengaluru_House_Data.csv')
df
```

**Check for NULL values**

```python
# Check for NULL values
df.isnull().sum()
```

## 2. **Address NULL values**
-------------------------------
Using Pandas, we can either drop that row, or fill it
> *Filling of values is called as Imputation*


Let's look at our dataset in more detail: 

- Location: Just one value -> Drop

- Size: Numerical -> 16 -> Drop

- Society: Categorical -> 5502 -> Fill

- Bath: Numerical -> 73 -> Fill

- Balcony: Numerical -> 609 -> Fill



### 2a. **Fill All Columns with common value**

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html

> **Fill all Null Values in the dataset with a single value**

```python
df2 = df.fillna(value = 0)
df2
```

```python
df2.isnull().sum()
```

> **Propogate previous observed value**

```python
df3 = df.fillna(method='ffill')
df3
```

> **Propogate next valid observation to current**

```python
df4 = df.fillna(method='bfill')
df4
```

### 2b. **Fill Only Categorical Column: Society**


**Update the entire table**

Option 1:
> df = df.fillna(value=True)

Option 2:
> df.fillna(value=True, inplace=True)


**Update only one column**

Option 1:
> df = df.fillna(value={'column1':value})

Option 2: 
> df.fillna(value={'column1':value}, inplace=True)


**Update only multiple chosen columns**

Both options exist. You can mention in the dict, all the list of columns to be updated

> df.fillna(value={'column1':value, 'column2':value2}, inplace=True)


```python
df5 = df.fillna(value={'society':'Unknown'})
df5
```

> **Fill numeric value**

```python
# For a categorical value, we can use mean (can't calculate), and median (isn't a meaningful replacement)
# The most suited statistic is 'mode'

values = {
    'society' : df['society'].mode()[0]
}

df6 = df.fillna(value=values)
df6
```

```python
df['society'].mode()
```

```python
df['society'].mode()[0]
```

> **bfill or ffill**

```python
df['society'].fillna(method='bfill', inplace=True)
df
```

### 2c. **Fill Multiple Numerical Columns: Bath, Balcony**


> **With Mean**

```python
# Fill with mean/median
values = {
    'bath' : df['bath'].mean(),
    'balcony': df['balcony'].mean()
}

df7 = df6.fillna(value=values)
df7
```

```python
df['bath'].mean()
```

```python
df7.isnull().sum()
```

> **With interpolate**

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html

```python
df6
```

```python
df6['balcony'] = df6['balcony'].interpolate(method='linear')
df6
```

### 2d. **Drop Other chosen columns**

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html


```python
df7
```

```python
df7.isnull().sum()
```

```python
df9 = df7.dropna(subset=['size', 'location'])
df9
```

```python
df9.isnull().sum()
```

# SUMMARY

1. The number of null rows determine if you have to drop or fill. 

**FILL**

2. Categorical Variables: 
- If the data is time-series, or has some sequential dependency: Fill with bfill, ffill
- If you want to put one blanket value: can use 'Mode', or a constant 'Unknown'

3. Numerical Variables:
- Can replace with single value
- Preferred: Replace with mean/median

4. Do a column-wise filling, rather than a common filling across the entire dataset (which might be quicker - good for v1, but, less accurate). 

**DROP**

5. If the number is insignificant (< 5%), you can drop. You don't want to drop a large part of your dataset.

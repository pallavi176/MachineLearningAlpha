**One-hot encoding**

For 'work_type', we have '5' possible values for the column.

One-hot encoding, replaces 'work_type' with 5 columns, each of which have 'True' or 'False'


Eg:
Old Column -> work_type = 'Private'

New Columns -> work_type_Private (with values 0 or 1)


> **5-columns**

```python
df = pd.get_dummies(data = df, columns = ['work_type'])
df
```

```python
# With drop first
df3 = pd.get_dummies(data = df, columns = ['gender'], drop_first=True)
df3
```
